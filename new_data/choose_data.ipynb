```python
# '보험계약'과 '청약서' 단어가 들어있는 사진 찾기
with open('/opt/ml/input/data/finance/ufo/finance_label.json') as reader:
    data = json.load(reader, encoding="UTF8")


cnt = 0
files = []
flag = False
flag2 = False
for idx,key in enumerate(data['images']):
    flag = False
    flag2 = False
    for i, x in enumerate(data['images'][key]['words']):
        if i > 100:
            break
        if (a:=data['images'][key]['words'][x]['transcription']) == "보험계약":
            flag=True
        if (b:=data['images'][key]['words'][x]['transcription']) == "청약서":
            flag2=True
        if flag and flag2:
            files.append(key)
            break

```

```python
files.sort()
```

```python
img_files = os.listdir('/opt/ml/input/data/finance/img')
for x in img_files:
    if x not in files:
        print(x)
        break
```

```python
# json 파일 만들기
new_dict = {}
for name in files:
    img_info = {name:
        data['images'][name]
    }
    new_dict.update(img_info)

result_json = {"images": new_dict}
with open('/opt/ml/input/data/finance/ufo/select_finance.json', 'w', encoding='UTF8') as writer:
    json.dump(result_json, writer, indent=4, ensure_ascii=False)
```
